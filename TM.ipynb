```python
import pandas as pd
df = pd.read_csv("Noboa.csv")
df
```

```python
df["ad_delivery_start_time"] = pd.to_datetime(df["ad_delivery_start_time"])
```

```python
#Ecuador had two election days in 2025. So we want to search for what happened during both previous months
range_1 = df["ad_delivery_start_time"].between("2025-01-09", "2025-02-09")
range_2 = df["ad_delivery_start_time"].between("2025-03-13", "2025-04-13")
```

```python
ecu_dates = df[range_1 | range_2]
ecu_dates
```

```python
#I want to check that the code ran correctly
print("Rango 1 -> min:", df[range_1]["ad_delivery_start_time"].min(),
      "| max:", df[range_1]["ad_delivery_start_time"].max())
```

```python
print("Rango 1 -> min:", df[range_2]["ad_delivery_start_time"].min(),
      "| max:", df[range_2]["ad_delivery_start_time"].max())
```

```python
#Now I want to know the money spent on that ads
#It will be difficult because the dataset hasn´t got exact numbers for spend
#So, I will count how many ads started between the two ranges
```

```python
res = df[range_1 | range_2]
len(res)
```

```python
res = df[range_1 | range_2]

percentage = len(res) / len(df) * 100
print(round(percentage, 2), "%")
```

```python
per_noboa = round(percentage, 2)
per_noboa
```
